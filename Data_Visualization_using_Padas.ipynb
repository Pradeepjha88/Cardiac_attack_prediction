# Draw Histogram

In Histogram

X axis is Values

Y axis is Frequency of that value (x)

What to Observe::

Distribution of data

1. If Numeric data:: See if it is like a bell curve

2. If Categorical Data :: check the balance of categories

```python
##import matplotlib.pyplot as plt # is compulsory
df[['V3']].hist() # V3, V21
```

```python
df[['V58']].hist() # Categories are almost balanced
```

```python
df[['V22']].hist() # Imbalance in categories is very high
```

# Draw Scatterplot

What to Observe ::

If line like pattern :: Very high correlation between X & Y

If Curve like pattern :: Some correlation between X & Y

If No pattern / all values are scattered :: NO correlation between X & Y

## NOTE:: Never do scatterplot for CATEGORICAL variables .. plots are not right

```python
df.plot.scatter(x='V21',y='V3')
```

```python
df.plot.scatter(x='V19',y='V3')
```

# Violin Plot

combination of Density plot and Box plot

```python
import seaborn as sb
```

## Single column violin

```python
sb.violinplot(x = "priceMin", data=df_ma)
plt.show()
```

# Two column Violin

X column should be categorical

Y column should be numeric

```python
# V4 is gender V3 is Age
sb.violinplot(x='V22',y='V3', data=df)
plt.show()
```

```python
sb.violinplot(x='V58',y='V3', data=df)
plt.show()
```
